#### We want to start by importing numpy's random module, to generate random integers for our list of values

```python
import numpy.random as random
from IPython.display import display
```

```python
# Will hold 1000 elements in the range 0 ~ 1000
values = [random.randint(1000) for x in range(1000)]
```

### So how would we go about finding a particular key value in the list?

#### Well, to be completely sure whether or not our key is in the list, we have to iterate through and ask:

    "Is the value at my current index equal to my key value (the value I'm looking for)?"

#### If yes: "Awesome! Return the index where I found my key value" 
#### If not: "Lame. Maybe it's at the next index?"

```python
# Let's take a look at our values list
values
```

```python
# Now we can do this three ways: the bad way, the right way, and the Pythonic way.
## The bad way is to hard-code a loop through our values list.
## Why is this bad? Because it isn't modular. We would have to rewrite each of these lines for
## any possible list we want to iterate over. We predefine our key by hand. That's lame.
key = 516
for value in values:
    if value == key: 
        print("Found it!")
```

```python
# Now let's do it the right way.
#
## We define a function called linear_search.
## What will the function need? We'll need an iterable object (i.e. a list, dictionary, etc.)
## and we'll need a key value to search for. So we can say:
def linear_search(iterable, key):
    # And just copy the above code into our linear search method,
    # making it modular and supporting good coding habits. Feels good.
    for value in iterable:
        if value == key:
            print("Found it!")
    print("Couldn't find it.")
```

```python
# Now let's do it the Pythonic way
#
## Python has this nifty inclusion operator called "in"
## Ergo, all we have to *really* do is say:
key in values
```

```python
## Now let's one-up the Pythonic way and modularize the "in" operator
## so that we can use any key, any iterable object, and still have it work
def linear_search(iterable, key):
    return key in iterable
```

```python
## Testing:
for x in range(len(values)):
    if linear_search(values, x):
        print(x)
```
